```python
df = pd.read_csv("data/ADAUSDT.30m.csv", names=traits.fields, usecols=["o", "h", "l", "c", "ot"], index_col=None)
#df.head()
df["ot"]
```

```python
date_time = pd.to_datetime(df["ot"], unit="ms")
```

```python
plot_cols = ["c", "o"]
plot_features = df[plot_cols]
plot_features.index = date_time
_ = plot_features.plot(subplots=True)
```
